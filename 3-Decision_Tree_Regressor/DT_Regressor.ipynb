# Decision Tree Regressor

## 1.Problem Statement

- The real estate market involves multiple factors such as house size, number of bedrooms, location, condition, and furnishing status that influence property prices. Accurately estimating house prices is essential for buyers, sellers, and real estate agents.

- The problem is to develop a predictive model using Decision Tree Regressor that can estimate the house price (house_price_usd) based on input features like:

- house size (sqft)
- number of bedrooms
- age of house
- location
- house type
- condition
- furnishing status

## 2.Objectives

- To analyze the relationship between house features and house prices
- To preprocess the dataset (handling categorical variables, scaling if needed)
- To build a Decision Tree Regressor model for predicting house prices
- To evaluate the model using metrics like:
     - Mean Squared Error (MSE)
     - R² Score
- To optimize model performance by tuning hyperparameters like:
     - max_depth
     - min_samples_split
     - min_samples_leaf

## 3.Import Libraries

```python
import numpy as np
import pandas as pd
```

## 4.Load Dataset

```python
df = pd.read_csv("ml_pipeline_practice.csv")
```

```python
df.head()
```

## 5.Check Missing Values

```python
df.isnull().sum()
```

## 6.Preprocessing

### i) One-Hot Encoding

```python
df = pd.get_dummies(df,columns=["location","house_type"]) # One-hot encoder
df
```

### ii) Label Encoding

```python
df["condition"] = df["condition"].map({
    "Poor": 0,
    "Average": 1,
    "Good": 2,
    "Excellent": 3
})

df["furnishing"] = df["furnishing"].map({
    "Unfurnished": 0,
    "Semi-Furnished": 1,
    "Furnished": 2
})
```

```python
df = df.astype(int)
```

```python
print(df.columns)
```

## 7.Model Building

### i) Seperate Target & Feature

```python
X = df.drop(["house_price_usd","id"],axis=1)
y = df["house_price_usd"]
```

### ii) Split The Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

### iii) Select Model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=3)
```

### iv) Train Model

```python
model.fit(X_train,y_train)
```

## 8.Predict

```python
y_pred = model.predict(X_test)
print(y_pred)
```

## 9.Evaluation

```python
from sklearn.metrics import mean_absolute_error,r2_score
MAE = mean_absolute_error(y_test,y_pred)
R2 = r2_score(y_test,y_pred)

print("MAE:",MAE)
print("R2:",R2)
```
