# 1.Problem Statement

- In today’s fitness industry, understanding customer behavior is crucial for improving subscription rates and user engagement. Gym and fitness app platforms collect various user data such as workout habits, app usage time, diet preferences, and fitness levels.

- The problem is to develop a classification model using Decision Tree Classifier that can predict whether a user will purchase a premium subscription (Subscription_Status) based on features like:

   - Age
   - Monthly Income
   - Workout Frequency
   - App Usage Time
   - Trial Days Used
   - Preferred Workout
   - Diet Type
   - Fitness Level
   - Device Type

- The goal is to assist businesses in identifying potential subscribers and improving targeted marketing strategies.

# 2.Objective

The main objectives of this project are:

- To analyze user behavior and identify patterns influencing subscription decisions
- To preprocess the dataset:
   - Handle categorical variables using encoding techniques
   - Prepare data for model training
- To build a Decision Tree Classifier model for predicting subscription status (0 or 1)
- To evaluate model performance using classification metrics:
   - Accuracy
   - Precision
   - Recall
   - F1 Score
- To understand the importance of features and decision rules generated by the model
- To reduce overfitting by tuning hyperparameters such as:
   - max_depth

## 3.Import Libraries

```python
import numpy as np
import pandas as pd
```

## 4.Load Dataset

```python
df = pd.read_csv("fitness_subscription_dataset.csv")
df.head()
```

## 5.Check Missing Values

```python
df.isnull().sum()
```

## 6.Preprocessing

### i) One-Hot Encoding

```python
df = pd.get_dummies(df,columns=["Preferred_Workout","Diet_Type"]) # One-hot encoder
df
```

### ii) Label Encoding

```python
df["Fitness_Level"] = df["Fitness_Level"].map({
    "Advanced":2,
    "Intermediate":1,
    "Beginner":0
})

df["Device_Type"] = df["Device_Type"].map({
    "iOS":1,
    "Android":0
})
```

## 7.Model Building

### i) Seperate Target & Feature

```python
X = df.drop(["Subscription","User_ID"],axis=1)
y= df["Subscription"]
```

### ii) Split The Dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

### iii) Select Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=6,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    random_state=42
)
```

### iv) Train Model

```python
model.fit(X_train,y_train)
```

## 8.Predict

```python
y_pred = model.predict(X_test)
print(y_pred)
```

## 9.Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report

print("accuracy_score:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred,output_dict=True))
```
